```python
from TCIAClient import TCIAClient
import pandas as pd
```

```python
base_url = 'https://services.cancerimagingarchive.net/services/v4/TCIA/query'
client = TCIAClient(base_url)
```

```python
collection_df = pd.DataFrame(data=client.get_json('getCollectionValues'))
collection_df.head()
```

```python
series_df = pd.DataFrame(data=client.get_json('getSeries', {'Collection': 'Lung-PET-CT-Dx'}))
series_df.head()
```

```python
patient_df = pd.DataFrame(data=client.get_json('getPatientStudy', {'Collection': 'Lung-PET-CT-Dx'}))
patient_df.head()
```

```python
test_uid = series_df.iloc[0]['SeriesInstanceUID']
```

```python
client.get_image(test_uid, './test/test.zip', False, False)
```
